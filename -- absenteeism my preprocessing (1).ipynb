```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv(r"C:\Users\hamza\OneDrive\Desktop\Absenteeism_data.csv")
raw_data.head()
```

```python
data = raw_data.copy()
data = data.drop('ID',axis = 1)
data
```

```python
len(data['Reason for Absence'].unique())
```

```python
data['Reason for Absence'].unique()
```

```python
data_without_reason = data.drop('Reason for Absence', axis = 1)
data_without_reason.shape
```

```python
data_with_dummies = pd.get_dummies(data['Reason for Absence'], prefix='Reason for Absence', drop_first=True)
data_with_dummies.shape
```

```python
data_with_dummies = data_with_dummies.drop('Reason for Absence_1', axis = 1)
data_with_dummies
```

```python
data_all = pd.concat([data_with_dummies, data_without_reason], axis = 1)
data_all.head()
```

```python
data_all['Date'] = pd.to_datetime(data_all['Date'], format='%d/%m/%Y')
data_all.head()
```

```python
data_all['Month'] = data_all['Date'].dt.month
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
data_all['Day of the week'] = data_all['Date'].apply(date_to_weekday)
```

```python
data_all = data_all.drop('Date', axis = 1)
```

```python
data_all.head()
```

```python
data_all['Education'].unique()
```

```python
data_all['Education'] = data_all['Education'].map({1:0,2:1,3:1,4:1})
```

```python
data_try = data_all.iloc[:,26:]
data_try.columns.values
```

```python
data.loc[3,'Reason for Absence']
```

```python
Reason_1 = []
Reason_2 = []
Reason_3 = []
Reason_4 = []
```

```python
for i in range(700):
    if 0<= data.loc[i,'Reason for Absence'] <=14:
        Reason_1.append(1)
    else:
        Reason_1.append(0)

        

        
for j in range(700):
    if data.loc[j,'Reason for Absence'] in [15,16,17]:
        Reason_2.append(1)
    else:
        Reason_2.append(0)
        
        
for k in range(700):
    if data.loc[k,'Reason for Absence'] in [18,19,20,21]:
        Reason_3.append(1)
    else:
        Reason_3.append(0)

        

for s in range(700):
    if data.loc[s,'Reason for Absence'] in [22,23,24,25,26,27,28]:
        Reason_4.append(1)
    else:
        Reason_4.append(0)
```

```python
len(Reason_4)
```

```python
Reason_df = pd.DataFrame({
                        'Reason_1':Reason_1,
                        'Reason_2':Reason_2,
                        'Reason_3':Reason_3,
                        'Reason_4':Reason_4
                        })
```

```python
Reason_df
```

```python
Data_preprocessed = pd.concat([Reason_df,data_try], axis = 1)
```

```python
Data_preprocessed
```

```python
Data_preprocessed.columns.values
```

```python
cols = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Month',
       'Day of the week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
Data_preprocessed = Data_preprocessed[cols]
```
